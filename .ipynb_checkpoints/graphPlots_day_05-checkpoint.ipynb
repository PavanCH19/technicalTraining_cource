```python
#univariable analysis
plt.figure(figsize = (10,6))
sns.distplot(df['city-mpg'], bins = 20)
plt.show()

#with border-color
plt.figure(figsize = (10,6))
#sns.distplot(df['city-mpg'], bins = 20, hist_kws = dict(edgecolor = '#ff0000'))
sns.histplot(df['city-mpg'], bins=20, kde=False, color='green', edgecolor='#ff0000', linewidth=2)
plt.show()
```

```python
#KDE plot
sns.set_style('darkgrid')
fig1, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (14,14))
x = np.random.normal(1,10,1000)

axes[0,0].set_title('Simple KDE Plot')
sns.kdeplot(x, ax = axes[0,0])

axes[0,1].set_title('Simple KDE Plot (Shaded Area under Curve)')
sns.kdeplot(x, ax = axes[0,1], shade=True)

axes[1,0].set_title('Simple KDE Plot (Different Color)')
sns.kdeplot(x, ax = axes[1,0], shade=True, color='r', cut=0)

axes[1,1].set_title('Simple KDE Plot (Density on veritcal axis)')
sns.kdeplot(x, vertical=True)
plt.show()
```

```python
#Countplot 
plt.figure(figsize=(6,9))
sns.countplot(y = 'body-style', data=df, order=df['body-style'].value_counts().index)
plt.show()
```

```python
#Box plot used to find outliers
sns.set_style(style='white')
plt.figure(figsize=(6,9))
sns.boxplot(df['engine-size'], )
```

```python
#multientries
plt.subplots(figsize=(14,24))
plt.subplot(3,1,1)
sns.boxplot(x=df['body-style'], y=df['engine-size'], hue=df['fuel-type'])
plt.subplot(3,1,2)
sns.boxplot(x=df['body-style'], y=df['wheel-base'], hue=df['fuel-type'])
plt.subplot(3,1,3)
sns.boxplot(x=df['body-style'], y=df['width'], hue=df['fuel-type'])
plt.show()
```

```python
# Boxen plots
plt.figure(figsize=(10,9))
sns.boxenplot(x=df['body-style'], y=df['engine-size'], palette='Set1')
plt.show()
```

Bivarite Analysis

```python
#Scatter plot
plt.figure(figsize=(10,10))
sns.scatterplot(x='engine-size', y='wheel-base', data=df, color='#ff0000', marker='o')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='engine-size', y='wheel-base', data=df, color='#ff0000', marker='o', hue='fuel-type', style='num-of-cylinders')
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(x='engine-size', y='wheel-base', data=df, hue='fuel-type', size='city-mpg', sizes=(20,200))
plt.show()
```

```python
#Line chart
sns.lineplot(data=df, x='engine-size', y='wheel-base')
plt.show()
```

```python
#Regplot - Regression plot
plt.figure(figsize=(9,9))
sns.regplot(x=df["compression-ratio"], y=df['engine-size'], color='#ff6600', logx=True)
plt.show()
```

```python
# Violin Plot
sns.set_style('white')
plt.figure(figsize=(16,8))
sns.violinplot(x=df['body-style'], y=df['city-mpg'], palette = 'coolwarm', scale='count')
plt.show()
```

Multivariate Analysis

```python
# Pairplot
sns.pairplot(data=df, hue='fuel-type', palette='husl', size=2)
plt.show()
```

```python
sns.pairplot(df, hue='fuel-type', x_vars=['wheel-base', 'length'], y_vars=['height','width'], markers=['o', 'D'], height=3, aspect=1)
plt.show()
```

```python
#Pairgrid
g = sns.PairGrid(df, hue='fuel-type', palette='husl')
g = g.map_offdiag(plt.scatter)
g = g.map_diag(plt.hist, histtype='step', linewidth = 2)
g = g.add_legend()
plt.show()
```

```python
# JointPlot
sns.set(style='white', color_codes=True)
g = sns.jointplot(x='wheel-base', y='engine-size', data=df, height=10, color='#ff6600')
plt.show()
```

```python
g = sns.jointplot(x='wheel-base', y='engine-size', data=df, height=10, kind='hex')
```

```python
g = sns.jointplot(x='wheel-base', y='engine-size', data=df, height=10, kind='kde', marginal_kws={'lw':5})
```

```python
# corr Matrix 
num_values = df.select_dtypes(include=[float, int])
corr = num_values.corr()
corr
```

```python
# heatmap
plt.figure(figsize=(20,10))
ax = sns.heatmap(corr, cmap='YlGnBu', linewidths=0.5, annot=True, fmt='.2f')
plt.yticks(rotation=0)
plt.show()
```
