```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2])
y = np.array([0, 1, 2])

# Create a 2x2 grid of subplots
plt.subplot(2, 2, 1)  # First subplot (top-left)
plt.plot(x, y, label="Line 1")
plt.title("Subplot 1")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.subplot(2, 2, 2)  # Second subplot (top-right)
plt.plot(x, y**2, label="Line 2", color="orange")
plt.title("Subplot 2")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.subplot(2, 2, 3)  # Third subplot (bottom-left)
plt.plot(x, np.sqrt(x), label="Line 3", color="green")
plt.title("Subplot 3")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.subplot(2, 2, 4)  # Fourth subplot (bottom-right)
plt.plot(x, np.exp(x), label="Line 4", color="red")
plt.title("Subplot 4")
plt.xlabel("X axis")
plt.ylabel("Y axis")

# Show the figure with all subplots
plt.tight_layout()
plt.show()
```
